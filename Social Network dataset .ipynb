```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv(r"C:\Users\Acer\Desktop\Dataset\Social_Network_Ads.csv")
```

```python
data.isnull().sum()
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold(n_splits=6,random_state=11,shuffle=True)
skf.get_n_splits(x)
print(skf)
```

```python
for train_index, test_index in skf.split(x,y):
    print("Train : ",train_index,"Test : ",test_index)
    x_train,x_test=x[train_index],x[test_index]
    y_train,y_test=y[train_index],y[test_index]
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.model_selection import cross_val_score
score=cross_val_score(lr,x_train,y_train,cv=skf)
print(score)
print(np.mean(score)*100)
```

```python
from sklearn.model_selection import cross_val_predict
y1_pred=cross_val_predict(lr,x_test,y_test,cv=skf)
print(y1_pred)
```
