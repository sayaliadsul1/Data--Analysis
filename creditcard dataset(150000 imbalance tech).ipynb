# Data Silicing

```python
x=data.iloc[:,0:28]
x
```

```python
y=data.iloc[:,-1]
y
```

# Data splitting into train & test 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print("Number transactions X_train dataset: ", x_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", x_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

# Train the model without handling the imbalanced class distribution

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train.ravel())
```

```python
y_pred=lr.predict(x_test)
y_pred
```

# Test size

```python
ab=lr.score(x_test,y_test)*100                  # accuracy : 99.89%
ab
```

# Imbalancing Techniques

# 1) Oversampling

```python
from imblearn.over_sampling import RandomOverSampler
ro=RandomOverSampler()
x_data,y_data=ro.fit_resample(x,y)
```

```python
from collections import Counter
```

```python
print("Actual data : ",Counter(y))
print("Artificial data : ",Counter(y_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train.ravel())
```

```python
y1_pred1=lr.predict(x_test)
```

```python
y1_pred1
```

# 2) Undersampling

```python
from imblearn.under_sampling import NearMiss
nm=NearMiss()
x1_data,y1_data=nm.fit_resample(x,y)
```

```python
print(x1_data.shape)
print(y1_data.shape)
```

```python
from collections import Counter
print("Actual data",Counter(y))
print("Artificial  data",Counter(y1_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1_data,y1_data,test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train.ravel())
```

```python
y2_pred=lr.predict(x_test)
y2_pred
```

# Test size

```python
C=d.iloc[[0,1,541,623],0:28]
C
```

```python
R=lr.predict(C)
R
```

```python
ab2=lr.score(x_test,y_test)*100                  # accuracy : 94.06%
ab2
```

# 3) SMOTE (Synthetic Minority Oversampling Technique)

```python
from imblearn.over_sampling import SMOTE
s=SMOTE()
x2_data,y2_data=s.fit_resample(x,y)
```

```python
print(x2_data.shape)
print(y2_data.shape)
```

```python
from collections import Counter
print("Actual data",Counter(y))
print("Artificial  data",Counter(y2_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x2_data,y2_data,test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train.ravel())
```

```python
y3_pred=lr.predict(x_test)
y3_pred
```
