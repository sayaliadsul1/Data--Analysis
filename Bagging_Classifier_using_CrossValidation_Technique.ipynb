```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

# Imbalancing technique (SMOTE)

```python
from imblearn.over_sampling import SMOTE
s=SMOTE()
x_data,y_data=s.fit_resample(x,y)
```

```python
from collections import Counter
print("The Original data : ",Counter(y))
print("The Aritificial data : ",Counter(y_data))
```

# Cross Validation Technique 

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,random_state=5,shuffle=True)
kf.get_n_splits(x_data)
print(kf)
```

```python
for train_index,test_index in kf.split(x_data):
    print("TRAIN : ",train_index,"TEST : ",test_index)
    x_train,x_test=x_data[train_index],x_data[test_index]
    y_train,y_test=y_data[train_index],y_data[test_index]
    
    from sklearn.linear_model import LogisticRegression
    lr=LogisticRegression()
    lr.fit(x_train,y_train)
    
    from sklearn.ensemble import BaggingClassifier
    bg=BaggingClassifier(base_estimator=lr,n_estimators=5,random_state=1)
    bg.fit(x_train,y_train)
    
    from sklearn.model_selection import cross_val_score
    from sklearn import model_selection
    results = model_selection.cross_val_score(bg, x_train, y_train, cv = kf)
    print("accuracy :")
    print(results)
    print(results.mean())
    
```
